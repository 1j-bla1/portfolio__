```python
%%javascript

//1.Person Class
class Person {
   constructor(name, age) {
       this.name = name;
       this.age = age;
   }
   //2.Greet
   greet() {
       console.log("Hello! My name is " + this.name);
   }
}

//3. Book Class
class Book {
   constructor(title, author, pages) {
       this.title = title;
       this.author = author;
       this.pages = Number(pages); // ensure it's a number
       this.pagesRead = 0;         // initialize pagesRead
   }
   //4. Pages
   read(pages) {
       let pagesToRead = Number(pages);
       this.pagesRead += pagesToRead;

       if (this.pagesRead > this.pages) {
           this.pagesRead = this.pages; // don't exceed total pages
       }

       console.log(`${pagesToRead} pages read. Total pages read: ${this.pagesRead}/${this.pages}`);
   }

   // Step 5: LongBook method
   isLongBook() {
       return this.pages > 300;
   }
}

// Step 6: Car class
class Car {
   constructor(make, model, fuel) {
       this.make = make;
       this.model = model;
       this.fuel = Number(fuel);
   }

   // Drive method
   drive(distance) {
       let fuelConsumed = distance * 0.1;
       if (fuelConsumed > this.fuel) {
           console.log(`${this.make} ${this.model} cannot drive ${distance} km. Not enough fuel.`);
       } else {
           this.fuel -= fuelConsumed;
           console.log(`${this.make} ${this.model} drove ${distance} km. Remaining fuel: ${this.fuel.toFixed(2)}`);
       }
   }

   // Refuel method
   refuel(amount) {
       this.fuel += Number(amount);
       console.log(`${this.make} ${this.model} refueled by ${amount}. Total fuel: ${this.fuel.toFixed(2)}`);
   }

   // Step 7: static method
   static info() {
       console.log("Cars are vehicles.");
   }
}

// Step 8: ElectricCar class
class ElectricCar extends Car {
   constructor(make, model, fuel, battery) {
       super(make, model, fuel);
       this.battery = Number(battery);
   }

   charge(amount) {
       this.battery += Number(amount);
       if (this.battery > 100) this.battery = 100;
       console.log(`${this.make} ${this.model} charged by ${amount}%. Battery level: ${this.battery}%`);
   }
}

// Test Person Class
let person1 = new Person("James", 14);

// Test Book Class
let myBook = new Book("The Hobbit", "J.R.R. Tolkien", 310);

// Printing Person Object
console.log("Person object:", person1);
console.log("Name:", person1.name);
console.log("Age:", person1.age);

// Printing Book Object
console.log("Book object:", myBook);
console.log("Title:", myBook.title);
console.log("Author:", myBook.author);
console.log("Pages:", myBook.pages);

// Printing Greet
person1.greet();

// Read some pages
myBook.read(50);
myBook.read(100);

// Return true if over 300 pages
console.log("Is this a long book?", myBook.isLongBook());

// Test the Car class
let myCar = new Car("Toyota", "Corolla", 50);
myCar.drive(100);
myCar.refuel(20);

// Static method print
Car.info();

// Output for electriccar
let myElectricCar = new ElectricCar("Tesla", "Model S", 0, 50);
myElectricCar.charge(30);
myElectricCar.charge(50);

// --- Step 10: Routine ---
function runRoutine() {
   console.log("===== Step 10: Numbered Combined Routine =====");

   // Create objects
   let person1 = new Person("James", 14);
   let myBook = new Book("The Hobbit", "J.R.R. Tolkien", 310);
   let myCar = new Car("Toyota", "Corolla", 50);
   let myElectricCar = new ElectricCar("Tesla", "Model S", 0, 50);

   // Step 1: Person Action
   console.log("1) Person Action:");
   console.log("   Name:", person1.name);
   console.log("   Age:", person1.age);
   person1.greet();

   // Step 2: Book Action
   console.log("2) Book Action:");
   console.log("   Title:", myBook.title);
   console.log("   Author:", myBook.author);
   console.log("   Pages:", myBook.pages);
   myBook.read(30);
   console.log("   Is this a long book?", myBook.isLongBook());

   // Step 3: Car Action
   console.log("3) Car Action:");
   myCar.drive(50);
   myCar.refuel(10);

   // Step 4: Electric Car Action
   console.log("4) Electric Car Action:");
   myElectricCar.charge(20);

   // Step 5: Static Car Info
   console.log("5) Car Class Info:");
   Car.info();

   console.log("===== End of Routine =====");
}

// --- Run the numbered routine ---
runRoutine();


// Outputing to dom

// Create a container under cell
const domOutput = document.createElement("div");
domOutput.style.background = "#f4f4f4";
domOutput.style.padding = "10px";
domOutput.style.marginTop = "20px";
domOutput.style.border = "2px solid #ccc";
domOutput.style.borderRadius = "10px";
domOutput.style.fontFamily = "monospace";
domOutput.style.whiteSpace = "pre-wrap";
domOutput.textContent = "Below is DOM Output:\n\n";
element.append(domOutput);

// Printing messages to both console and DOM
function logToDom(...messages) {
    const text = messages.map(m =>
        typeof m === "object" ? JSON.stringify(m, null, 2) : m
    ).join(" ");
    console.log(...messages);
    const p = document.createElement("p");
    p.textContent = text;
    domOutput.appendChild(p);
}

// Run DOM
function runRoutineDOM() {
    logToDom("===== Step 10: Numbered Combined Routine (DOM Output) =====");

    let person1 = new Person("James", 14);
    let myBook = new Book("The Hobbit", "J.R.R. Tolkien", 310);
    let myCar = new Car("Toyota", "Corolla", 50);
    let myElectricCar = new ElectricCar("Tesla", "Model S", 0, 50);

    logToDom("1) Person Action:");
    logToDom("   Name:", person1.name);
    logToDom("   Age:", person1.age);
    logToDom("   Greeting: Hello! My name is " + person1.name);

    logToDom("2) Book Action:");
    logToDom("   Title:", myBook.title);
    logToDom("   Author:", myBook.author);
    logToDom("   Pages:", myBook.pages);
    myBook.read(30);
    logToDom("   Is this a long book?", myBook.isLongBook());

    logToDom("3) Car Action:");
    myCar.drive(50);
    myCar.refuel(10);

    logToDom("4) Electric Car Action:");
    myElectricCar.charge(20);

    logToDom("5) Car Class Info:");
    logToDom("Cars are vehicles.");

    logToDom("===== End of Routine =====");
}

// Run DOM output version
runRoutineDOM();

//I hope you enjoyed my work. This took a lot of time to make. Thanks
```
